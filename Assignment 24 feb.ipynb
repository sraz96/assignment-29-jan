```python
 Here are five functions of the pandas library with examples of their execution:
```

```python
1.read_csv(): This function is used to read data from a CSV file and create a pandas DataFrame object.
```

```python
df = pd.read_csv('services.csv')
print(df.head())
```

```python
2.groupby(): This function is used to group data in a DataFrame based on one or more columns and perform operations on the groups.
```

```python
data = {
    'name': ['Alice', 'Bob', 'Claire', 'David', 'Emily'],
    'age': [25, 30, 27, 32, 28],
    'gender': ['Female', 'Male', 'Female', 'Male', 'Female']
}

df = pd.DataFrame(data)

grouped = df.groupby('gender')['age'].mean()

print(grouped)
```

```python
3.dropna(): This function is used to remove rows or columns from a DataFrame that contain missing values.
```

```python
df = pd.read_csv('services.csv')
clean_df = df.dropna()
print(clean_df.head())
```

```python
4.concat(): This function is used to concatenate two or more pandas objects, such as DataFrames or Series, along a particular axis.
```

```python
df1 = pd.read_csv('services.csv')
df2 = pd.read_csv('services1.csv')
combined_df = pd.concat([df1, df2], axis=0)
print(combined_df.head())
```

```python
5.pivot_table(): This function is used to create a pivot table from a DataFrame, where the values are aggregated based on one or more columns.
```

```python
df = pd.read_csv('example.csv')
pivot_table = pd.pivot_table(df, values='Price', index='Category', columns='Store', aggfunc='mean')
print(pivot_table.head())
```

```python
def count_words(df):
    df['Word_Count'] = df['Text'].apply(lambda x: len(x.split()))
    return df
```

```python
df = pd.DataFrame({'Text': ['The quick brown fox', 'jumped over the lazy dog']})
df = count_words(df)
print(df)
```

```python
DataFrame.size() and DataFrame.shape() are both functions in the Pandas library, but they return different information about a DataFrame:

DataFrame.size() returns the total number of elements in a DataFrame, which is equivalent to the number of rows multiplied by the number of
columns. It returns a scalar value, not a tuple.
DataFrame.shape() returns a tuple of two integers representing the number of rows and columns in a DataFrame, respectively. It does not count 
the number of elements, but rather provides the dimensions of the DataFrame.
For example, if a DataFrame has 10 rows and 5 columns, DataFrame.size() would return 50 and DataFrame.shape() would return (10, 5).
```

```python
def select_rows(df):
    selected_df = df[(df['A'] > 5) & (df['B'] < 10)]
    return selected_df
```

```python
def add_moving_average(df):
   
    df = df.sort_values('Date')
    df['MovingAverage'] = df['Sales'].rolling(window=7, min_periods=1).mean()

    return df
```

```python
df = pd.DataFrame({
    'Date': pd.date_range(start='2022-01-01', end='2022-01-31'),
    'Sales': [10, 15, 8, 12, 7, 9, 20, 18, 22, 17, 16, 25, 30, 20, 28, 32, 35, 40, 45, 50, 35, 28, 22, 18, 15, 10, 5, 8, 12, 16]
})

df = add_moving_average(df)

print(df)
```

```python
import pandas as pd

def add_weekday_column(df):
    df['Date'] = pd.to_datetime(df['Date'])
    df['Weekday'] = df['Date'].dt.weekday_name
    return df
```

```python
df = pd.DataFrame({'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05']})

df['Date'] = pd.to_datetime(df['Date'])
df['Weekday'] = df['Date'].dt.day_name()

print(df)
```

```python
def select_date_range(df):
    start_date = '2023-01-01'
    end_date = '2023-01-31'
    mask = (df['Date'] >= start_date) & (df['Date'] <= end_date)
    filtered_df = df.loc[mask]
    return filtered_df
```

```python
filtered_df = select_date_range(df)
print(filtered_df)
```
